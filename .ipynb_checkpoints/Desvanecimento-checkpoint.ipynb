## Espalhamento Doppler

Quando um sinal senoidal de frequência $f_c$ é transmitido, o espectro do sinal recebido denominado de **Espectro Doppler** tem componentes entre $f_c - f_d$ e $f_c + f_d$, sendo $f_d$ o deslocamenteo Doppler, que é função da velocidade do móvel $v$ e do ângulo $\theta$ entre a estação móvel e a direção de chegada das ondas dispersas.

Se a largura de banda do sinal for muito maior que $B_d$, então o o efeito do espalhamento Doppler no receptor é desprezível, e o canal é dito ser de atenuação lenta.

O tempo de coerência do canal é a duração de tempo no qual a resposta ao impulso do canal não varia. 




Seja $$h(t) = \sum_{n=1}^M a_i\delta(t - T_i)$$ um canal com $M$ percursos, cada um com atenução $a_i$ e atraso $T_i$.

**Exemplo** para $M = 2$ temos $$h(t) = a_1\delta(t- T_1) + a_2\delta(t- T_2)$$ e aplicando a Transformada de Fourier chegamos a 
$$ H(\omega) = a_1e^{-j\omega T_1} + a_2e^{-j\omega T_2}$$ que tem módulo
$$|H(\omega)| = \sqrt{1 + k^2 + 2k\cos(\omega \Delta T)}$$ sendo $\Delta T = T_2 - T_1$ e $k = a_2/a_1$, que está desenhado pelo código a seguir

```python
import numpy as np
import matplotlib.pyplot as plt

k = 1

Delta_t = 1

omega = np.linspace(0,10*np.pi,512)

AbsH = np.sqrt(1 + k*k + k*np.cos(omega*Delta_t))
               
plt.plot(AbsH)
```

Se a largura de banda do sinal transmitido nesse canal é menor $\omega\Delta T$, então o canal é dito ser de **banda estreita**, e nesse caso a atenuação em frequência é uniforme.

Caso contrário, a banda do sinal é maior que $\omega\Delta T$ o canal é dito de **banda larga** a atenução não é uniforme e é seletiva em frequência.

Até o momento consideramos que a função de transferência não muda com o tempo, porém se o canal varia com o tempo, pode-se definir o **Tempo de Coerência** ($T_c)$  como a mediada de duração em que o canal permanece constante.

Na frequência, a medida de alargamento causada pelo mudança do canal é denominada de **Espalhamento Doppler**. Por exemplo, uma senóide com frequência $f_c$ no repcetor tem componentes entre $f_c - f_m$ e $f_c + f_m$. Essa variação lembra a modulação em frequência.
