## Demodulação

O sinal produzido na saída do canal entra no receptor que realizará o processamento para recuperar a mensagem transmitida. A primeira etapa no processo de recepção é realizar a demodulação, que consiste em transladaro escpetro da portadora para a banda básica, ou frequência intermidiária, o que implica em recuperar o sinal mensagem.

A demodulação pode ser coerente quando informações sobre a fase e a ferquência da portadora são usados para recuperar mensagem, ou não coerente quando não se requer sincronização entre a frequência e fase da portadora.

### Rádio QAM

Para aumentar a eficiência na utlização da banda em sinais modulados em amplitude pode-se transmitir um sinal em fase ($\cos\omeg_ct) e outro sinal em quedratura ($\sin\omega_c t$), obtendo assim o sinal em quadratura
\[
q(t) = m_1(t)\cos(\omega_c t + \theta) - m_2(t)\sin(\omega_ct + \theta).
\]
Quando a informação está no formato digital, vimos que devemos formtar um pulso obtendo um sinal que represente aquela informação, ou seja, dado que $s_i(k)$ é a sequência de símbolos da $i$-ésima mensagem temos
\[
m_i(t) = \sum_{l}s_i(k)p(t - kT)
\]
sendo $p(t)$ o pulso escolhido. Assim o sinal transmitido está modulado em quadratura e pode ser escrito como
\[
s(t) = \sum_kp(t - kT)[s_1(k)\cos(\omega_ct + \theta) - s_2(k)\sin(\omega_c t + \theta)].
\]
Como exemplo, seja o par $(s_1, s_2) \in \{(1,1), (1,-1), (-1,1), (-1,-1)\}, então pode-se associar dois bits a cada combinação e transmitir 2 bits por intervalo de transmissão, nesse caso temos o sinal QAM. O mesmo pode ser coneguido via uma modulação em fase considerando as fases $\{\pi/4, 3\pi/4, 5\pi/4, 7\pi/4\}$, nesse caso temos o sinal QPSK.

### Costas Loop

A recepção quando perfeitamente sincronizada vai produzir os pontos da constelação do QAM, entretanto como vimos pode haver rotação na constelação, e dependendo dessa rotação poderá haver ambiguidade na recepção. 

O objetivo é ajustar o misturador no receptor local para garantir que $\phi(t) = 2\pi(f_c - f_0)t + \theta  +\pho\pi/2$ quando $\pho$ é fixo. Assumindo que $f_c = f_0$ o objeivo passa a ser $\phi = \theta + \pho(\pi/2)$.
